```python
colnames = ['bairro', 'zona', 'local', 'endereco', 'secao1', 'secao2']
zona_sp = pd.read_csv('/Users/willianribeiro/Documents/GitHub/dados-eleicoes-brasil-municipios/data/zonas/arquivo.csv', names=colnames, encoding='Latin 1')
```

```python
zona_sp.head()
```

```python
endereco = zona_sp.iloc[:,0:4]
```

```python
endereco.head()
```

```python
endereco['endereco_formatado'] = endereco['local'] + ', ' + endereco['endereco']
endereco.head()
```

```python
ponto = geocode(endereco['endereco_formatado'], provider="nominatim", timeout=None, user_agent='test_application')
```

```python
uniao = ponto.merge(endereco, left_index=True, right_index=True)
```

```python
uniao.head()
```

```python
remover = uniao[uniao['geometry'].is_empty]
remover['endereco_formatado'] = remover['zona'] + ', SÃO PAULO'
remover
```

```python
rem_ponto = geocode(remover['endereco_formatado'], provider="nominatim", timeout=None, user_agent='test_application')
```
